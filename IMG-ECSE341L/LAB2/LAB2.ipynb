```
image = cv2.imread('img1.png')
image2=cv2.imread('img2.png')
```

```
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```
cv2_imshow(image)
print('Converting RGB to GREYSCALE using OpenCV')
cv2_imshow(gray)
```

```
import matplotlib.image as mpimg
img=cv2.imread('img2.png')
b, g, r = img[:,:,0], img[:,:,1], img[:,:,2]
gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
plt.imshow(gray,cmap=plt.get_cmap('gray'))
```

```
print('Performed Addition, subtraction and multplication on 2 images')
sub,add,mup = image-image2, image+image2, image*image2
imgplot = plt.imshow(sub)
plt.show()
imgplot = plt.imshow(add)
plt.show()
imgplot = plt.imshow(mup)
plt.show()
```

```
from matplotlib import colors
img_hsv = colors.rgb_to_hsv(image)
imgplot = plt.imshow(img_hsv)
plt.show()
print("CONVERTED IMG FROM RGB TO HSV ")
```

```
org_scale = PIL.Image.open('img1.png')
width, height = org_scale.size  
print("Width :",width,"\nHeight: ",height)
#performing cropping operations
cropped = org_scale.crop((4, height/5, 200, 3 * height / 5)) 
cwidth, cheight = org_scale.size  
newsize = (600,613) #2x factor
bilinearimg = cropped.resize(newsize,resample=PIL.Image.BILINEAR)
bicubicimg = cropped.resize(newsize,resample=PIL.Image.BICUBIC)
NNimg = cropped.resize(newsize,resample=PIL.Image.NEAREST)
print("Bilinear")
imgplot = plt.imshow(bilinearimg)
plt.show()
print("BiCubic")
imgplot = plt.imshow(bicubicimg)
plt.show()
print("Nearest Neighbour")
imgplot = plt.imshow(NNimg)
plt.show()
```
