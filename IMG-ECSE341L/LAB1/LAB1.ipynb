```
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

from IPython.display import display
```

```
# Loading Images using Python Imaging Library(PIL)
path='/content/drive/My Drive/BTECH SEM 5/Image&VideoProcessing/img.jpg'
load_img=Image.open(path)
display(load_img)
```

```
# Loading Images using MATPLOTLIB
path='/content/drive/My Drive/BTECH SEM 5/Image&VideoProcessing/img.jpg'
img=plt.imread(path)
plt.imshow(img)
plt.show()
```

```
# Loading Images using CV2
img=cv2.imread(path)
cv2_imshow(img)
```

```
#loading image with skimage

from skimage.io import imread,imshow

image = imread(path)
imshow(image)
```

```
#Loading image without using any library except Numpy
ROWS=512
COLS=512
imgread=open('/content/drive/My Drive/BTECH SEM 5/Image&VideoProcessing/barbara_gray.raw')
img = np.fromfile(imgread, dtype = np.uint8, count = ROWS * COLS)
img.shape = (img.size // COLS, COLS)
print("... Save the output image")
print(img)
img.astype('int8').tofile('/content/drive/My Drive/BTECH SEM 5/Image&VideoProcessing/NewImage.raw')
print("... File successfully saved as NewImage.raw")
imgread.close()
```
